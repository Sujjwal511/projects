```python
import pandas as pd    #importing pandas lib
import numpy as np     #importing numpy lib
import seaborn as sns  
import matplotlib.pyplot as plt 
%matplotlib inline

#reading CSV file 
df = pd.read_csv('SaudiWeather.csv')
df  
```

```python
df.head()
```

```python
df.info()
```

```python
#convert humidity to numerical
df['humidity'] = df['humidity'].str.replace('%',' ').astype('float')
```

```python
df.isnull().sum()
```

```python
#mean as per city
df.groupby(by='city')[['humidity','barometer']].mean().reset_index()
```

```python
df.describe()
```

```python
#unique weather
df['weather'].unique()
```

```python
#only related to rain weather we will store in rain var.
df['rain'] = df['weather'].str.contains('rain|shower|thunderstorm|thundershowers', case = False ,regex=True)
df
```

```python
#check for sample all the weather type is under rain showing true or not.
df.groupby('weather')['rain'].unique().sample(20)
```

```python
#values count in rain var.
df['rain'].value_counts()
```

```python
#drop weather column as we dont need it not
df.drop('weather', axis =1 , inplace =True)
df
```

```python
plt.figure(figsize=(15,10)) #size of the fig
sns.scatterplot(x='humidity', y='temp',hue='rain',palette='deep',data=df) 
```
