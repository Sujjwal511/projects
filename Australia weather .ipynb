```python
#To predict tmrw it will rain or not in australia.
import pandas as pd     
import numpy as np    
import seaborn as sns  
import matplotlib.pyplot as plt 
%matplotlib inline

df = pd.read_csv('AustraliaWeather.csv')
df.head()
```

```python
#information ofthe dataset
df.info()
```

```python
#checking for null values
df.isnull()
```

```python
#to check how many null values are there in column
df.isnull().sum()
```

```python
#dropping the columns which we dont need.
df=df.drop(['Evaporation','Evaporation','Cloud9am','Cloud3pm','Location','Date'], axis=1)
df.head()
```

```python
#dropping the rows in the dataset which has null values.
df=df.dropna(axis=0)
df.shape
```

```python
#name of the columns in dataset in which raintmrw wich need to be predicted.
df.columns
```

```python
#by using label encoding we need to convert string into numerical data.
from sklearn.preprocessing import LabelEncoder #imprted labelencoder from sklear.preprocessing
le = LabelEncoder()                            #
df['WindGustDir']=le.fit_transform(df['WindGustDir']) #transform into numerical and store in ame column.
df['WindDir9am']=le.fit_transform(df['WindDir9am'])    #
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])     #
df['RainToday']=le.fit_transform(df['RainToday'])       #
df['RainTomorrow']=le.fit_transform(df['RainTomorrow'])   #
```

```python
df.head()
```

```python
#divide datainto dependent and independent data. Updating x and y.
x=df.drop(['RainTomorrow'],axis=1)  #dependent column. 
y=df['RainTomorrow']       #need to be predicted.
```

```python
#lets check in x last column is not in cluded
x.head()
```

```python
#as min temp is incresing the max temp is increasing so it is leanear form
plt.figure(figsize=(15,10)) #size of the fig
sns.scatterplot(x='MaxTemp', y='MinTemp',hue='RainTomorrow',palette='bright',data=df) #maxtemp vs mintemp plot on scatterplt.
```

```python
#plot regarding humidity and temp.
plt.figure(figsize=(15,10))
sns.scatterplot(x='Humidity9am', y='Temp9am', hue='RainTomorrow',palette='colorblind',data=df)
```

```python
#heatmap to show the corellation of rain tmrw.
#good realtionship with humidity9am and humidity3pm. and raintoday also.
plt.figure(figsize=(15,10))
sns.heatmap(df.corr())
```

```python
from sklearn.model_selection import train_test_split #imported train test to split data.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3) #30% of data from the dataset.
```

```python
from sklearn.metrics import accuracy_score
```

```python
#logistic regresstion.
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()   #creating object of it.
lr.fit(x_train,y_train)   #fitting xtrain and ytrain.
predictions=lr.predict(x_test)  #what to be predicted.
print(accuracy_score(y_test,predictions))  #accuracy is 85%
```

```python
#imported required libraries. CSV file also
import pandas as pd     
import numpy as np    
import seaborn as sns  
import matplotlib.pyplot as plt 
%matplotlib inline

df = pd.read_csv('AustraliaWeather.csv')
df.head()
```

```python
#shape of the dataset
df.shape
```

```python
#first 5 rows of the dataset
df.head()
```

```python
#returns the number of missing values in each column
df.isna().sum()
```

```python
#dropping all null values and shape of the dataset
df = df.dropna()
df.shape
```

```python
#returns the number of missing values in each column
df.isna().sum()
```

```python
#getting data in the particular area.
df = df[df['Location']=='Darwin']
df.shape
del df['Location']
```

```python
#information of the current dataset
df.info()
```

```python
#description of the current dataset
df.describe()
```

```python
#data visualization. 
#getting month from date.
def get_month(date):  #get the particular column which we want
    return int(date.split('-')[1]) #split the date
df = df.rename(columns={'Date': 'Month'})   #rename colum to month 
df['Month'] = df['Month'].apply(get_month) #keep only month in column month.
```

```python
#month colum added in the dataset
df.head()
```

```python
#over all rain according to month.
#plt.figure(figsize=(10,5))
sns.countplot(data = df, x='Month', hue = df['RainTomorrow'])
plt.show()
```

```python
#it will rain tmrw depend on rain today prediction
plt.figure(figsize=(10,5))
sns.countplot(data = df, x = 'RainToday', hue=df['RainTomorrow']);
```

```python
#rain tmrw or not according to windgustdir
plt.figure(figsize=(10,5))
sns.countplot(data = df, x = 'WindGustDir', hue=df['RainTomorrow']);
```

```python
#rain tmrw as per the wind at 3pm
plt.figure(figsize=(10,5))
sns.countplot(data = df, x = 'WindDir3pm', hue=df['RainTomorrow']);
```

```python
#rain tmrw as per the wind at 9am
plt.figure(figsize=(10,5))
sns.countplot(data = df, x = 'WindDir9am', hue=df['RainTomorrow']);
```

```python
#max temp vs min temp. as temp will decrease it will rain else no.
plt.figure(figsize=(10,5))
sns.scatterplot(x='MaxTemp', y='MinTemp', data=df, hue='RainTomorrow');
```
